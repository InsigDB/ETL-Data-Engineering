# Create Functions For Clean Data Sources If Neccassery 

```python
def clean_clients(df):# create clean_clients function  for cleaning df_clients 
    df = df.drop_duplicates()
    df = df.drop(columns = 'Unnamed: 0')
    df = df.rename (columns = {'client_id': 'client_id' , 'nom' :'last_name' , 'prenom' :'first_name'  , 'ville_id' :'city_id' })
    return df
```

```python
def clean_villes(df):
    df = df.drop_duplicates()
    df = df.rename (columns = {'ville_id': 'city_id' , 'nom_ville' :'city_name' , 'pays_id' :'country_id' })
    return df
```

```python
def clean_pays(df):
    df = df.drop_duplicates()
    df = df.rename (columns = {'pays_id' : 'country_id' , 'nom_pays' : 'country_name'  })
    return df
```

```python
def clean_compts(df):
    df = df.drop_duplicates()
    df = df.rename (columns = {'compte_id': 'account_id' , 'client_id' :'client_id' , 'type_compte' :'account_type'  , 'date_douverture' : 'opening_date' })
    return df
```

```python
def clean_transactions(df):# 
    df = df.drop_duplicates() 
    df = df.rename (columns = {'transaction_id': 'transaction_id' , 'compte_id' :'account_id' , 'date_id' :'date_id' , 'montant' :'amount' , 'type_transaction' :'transaction_type' })
    return df
```

```python
def clean_dates(df):
    df = df.drop_duplicates() 
    df = df.rename (columns = {'date_id': 'date_id' , 'jour' :'day' ,  'mois':'month', 'trimestre' :'quarter' , 'annee' : 'year' })
    return df
```

# Applying functions 

```python
df_clean_clients = clean_clients(df_clients) #  applying function 
df_clean_clients 
```

```python
df_clean_villes = clean_villes(df_villes)
df_clean_villes
```

```python
df_clean_pays = clean_pays(df_pays)
df_clean_pays 
```

```python
df_clean_compts = clean_compts(df_comptes)
df_clean_compts
```

```python
df_clean_transactions = clean_transactions(df_transactions)
df_clean_transactions
```

```python
df_clean_dates = clean_dates(df_dates)
df_clean_dates
```
